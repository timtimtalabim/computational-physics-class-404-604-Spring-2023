```python
import numpy as np
import matplotlib.pyplot as plt

def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx

def derivative_test(der, x):
    f = lambda i: np.cos(i)

    dx = [1.0*10**i for i in range(-16,0)]

    f_prime = lambda i: -1*np.sin(i)

    f_numeric = [der(f, x, i) for i in dx]

    f_analytic = f_prime(x)

    f_error = [abs(i-f_analytic) for i in f_numeric]

    for i, j in zip(dx, f_error):
        print(f"dx: {i}, error: {j}")
    print(f"best error: {min(f_error)} at dx: {dx[f_error.index(min(f_error))]}")
    plt.plot(dx, f_error)
    plt.xscale('log')
    plt.yscale('log')
    
derivative_test(derifw, np.pi/4)

# Analytically it seems like dx^n, n is equal to 2
# Optimal value will be the minimum of this graph which is 10^-8
```

```python
def dericd(y,x,dx):
    return (y(x+(dx/2))-y(x-(dx/2)))/dx

derivative_test(dericd, np.pi/4)

# Analytically it seems like dx^n, n is equal to 2
# Optimal value will be the minimum of this graph which is 10^-5
```

```python
derivative_test(dericd, 0)

# Analytically this makes sense, since x at 0 is an extrema, the value on either side of
# this extrema will be the same and return 0.
```
