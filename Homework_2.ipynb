```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, n):
    
    fsum = 0
    for i in range(1, n+2, 2):
        fsum += np.sin(i*x)/i
    
    final = np.pi/2 + (2*fsum)
    
    return final

print(f(np.pi, 11))

x = np.linspace(-np.pi, np.pi)

y1 = f(x, 11)
y2 = f(x, 21)
y3 = f(x, 31)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```
